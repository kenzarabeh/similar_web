```python
# Extraction des donn√©es sites web - Test sur amazon.fr et joueclub.fr
import time

def extract_website_data(domain, start_month, end_month, country='fr'):
    """
    Extraction des 6 m√©triques pour un site web
    """
    print(f"\nüåê Extraction pour {domain}")
    
    # Endpoints pour les 6 m√©triques
    metrics_endpoints = {
        'visits': '/total-traffic-and-engagement/visits',
        'pages_per_visit': '/total-traffic-and-engagement/pages-per-visit',
        'avg_visit_duration': '/total-traffic-and-engagement/average-visit-duration', 
        'bounce_rate': '/total-traffic-and-engagement/bounce-rate',
        'page_views': '/total-traffic-and-engagement/page-views',
        'desktop_mobile_split': '/total-traffic-and-engagement/desktop-vs-mobile'
    }
    
    # Param√®tres communs
    params = {
        'start_date': start_month,
        'end_date': end_month,
        'country': country,
        'granularity': 'monthly',
        'main_domain_only': 'false',
        'format': 'json'
    }
    
    domain_results = {
        'domain': domain,
        'metrics': {}
    }
    
    for metric_name, endpoint in metrics_endpoints.items():
        print(f"  üìä Extraction {metric_name}...")
        
        # Construction de l'endpoint complet
        full_endpoint = f'/website/{domain}{endpoint}'
        
        # Appel API
        data = make_api_request(full_endpoint, params)
        
        if data:
            print(f"    ‚úÖ {metric_name} r√©cup√©r√©")
            domain_results['metrics'][metric_name] = data
        else:
            print(f"    ‚ùå √âchec {metric_name}")
            domain_results['metrics'][metric_name] = None
        
        # Pause entre requ√™tes
        time.sleep(1)
    
    return domain_results

# Configuration
DOMAINS = ['amazon.fr', 'joueclub.fr']
START_MONTH = '2024-04'
END_MONTH = '2024-04'

print("üöÄ Extraction des donn√©es sites web")
print(f"üìÖ P√©riode : {START_MONTH}")
print(f"üåç Pays : fr")
print(f"üí∞ Co√ªt : {len(DOMAINS)} domaines √ó 6 m√©triques = {len(DOMAINS) * 6} cr√©dits")

EXECUTE_WEBSITES = True  # Changez √† True pour lancer

if EXECUTE_WEBSITES:
    print("\nüöÄ Lancement de l'extraction sites web...")
    
    websites_results = []
    
    for domain in DOMAINS:
        result = extract_website_data(domain, START_MONTH, END_MONTH, 'fr')
        websites_results.append(result)
    
    print(f"\nüéâ Extraction termin√©e ! {len(websites_results)} domaines trait√©s")
    
    # Sauvegarde
    import json
    with open('../data/extraction_sites_web_avril2024.json', 'w', encoding='utf-8') as f:
        json.dump(websites_results, f, indent=2, ensure_ascii=False)
    
    print("üíæ Donn√©es sauvegard√©es : data/extraction_sites_web_avril2024.json")
    
else:
    print("‚è∏Ô∏è Extraction en attente (EXECUTE_WEBSITES = False)")
    print("\nüí° Pour lancer l'extraction, changez EXECUTE_WEBSITES = True")
```
